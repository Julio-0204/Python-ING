## Sistema Basico de Robots

```python
#------------Parte 1------------------------------------------------------------
class Robot:
    def __init__(self, name, model):
        self.name = name
        self.model = model
        self.battery = 100.0  # Batería inicial al 100%
        self.distance_traveled = 0.0  # Distancia inicial 0 km
        self.speed = 0.0  # Velocidad inicial 0 km/h
        self.direction = 0  # Dirección inicial 0 grados
        print(f"Robot {self.name} modelo {self.model} creado")
#-------------------------------------------------------------------------------
#informacion del robot
    def __repr__(self):
        return f"{self.name} ({self.model}) - Batería: {self.battery}% - Distancia: {self.distance_traveled}km"

 #------------------------------------------------------------------------------
 #mueve el robot en una direccion en un tiempo determiinado
    def move(self, speed, direction, time_minutes):
        self.speed = speed
        self.direction = direction

 #******************************************************************************
        #calcular distancia recorrida
        distance = speed * (time_minutes / 60)
        self.distance_traveled += distance

        #reducir batería (10% por cada km)
        battery_consumed = distance * 10
        self.battery -= battery_consumed

        #verificar si la bateria esta baja
        if self.battery < 20:
            print("alerta bateria baja")

        print(f"{self.name} se movió {distance}km en dirección {direction}°")

    def charge(self, time_minutes):
        #aumentar batería (2% por minuto)
        battery_increase = time_minutes * 2
        self.battery += battery_increase

        #maximo 100%
        if self.battery > 100:
            self.battery = 100.0
 #******************************************************************************

 #-------------------Estadisticas parte 1---------------------------------------
        print(f"{self.name} cargando... batería: {self.battery}%")            #-
    def get_status(self):                                                     #-
        """Muestra información completa del robot"""                          #-
        print(f"\n=== Estado de {self.name} ===")                             #-
        print(f"Modelo: {self.model}")                                        #-
        print(f"Batería: {self.battery}%")                                    #-
        print(f"Distancia total: {self.distance_traveled}km")                 #-
        print(f"Velocidad actual: {self.speed}km/h")                          #-
        print(f"Dirección actual: {self.direction}°")                         #-
#-------------------------------------------------------------------------------

#--------------------Parte 2----------------------------------------------------

class WorkerRobot(Robot):
    def __init__(self, name, model):
        super().__init__(name, model)
        self.cargo_weight = 0.0
        self.max_cargo = 100.0

    def load_cargo(self, weight):
        if self.cargo_weight + weight <= self.max_cargo:
            self.cargo_weight += weight
            print(f"{self.name} cargó {weight}kg. Carga total: {self.cargo_weight}kg")
        else:
            print(f"{self.name} no puede cargar {weight}kg. Capacidad máxima excedida")

    def unload_cargo(self):
        if self.cargo_weight > 0:
            print(f"{self.name} descargó {self.cargo_weight}kg")
            self.cargo_weight = 0.0
        else:
            print(f"{self.name} no tiene carga para descargar")

    def move(self, speed, direction, time_minutes):
        if self.cargo_weight > 50:
            speed = speed * 0.5
            print(f"Velocidad reducida por carga pesada: {speed}km/h")
        super().move(speed, direction, time_minutes)

    def get_status(self):
        super().get_status()
        print(f"Carga actual: {self.cargo_weight}kg / {self.max_cargo}kg")

class GuardRobot(Robot):
    def __init__(self, name, model):
        super().__init__(name, model)
        self.alert_mode = False  #modo alerta inicialmente desactivado

#----------------modo alerta carga 50% más lento--------------------------------
    def charge(self, time_minutes):
        if self.alert_mode:
            battery_increase = time_minutes * 1
        else:
            battery_increase = time_minutes * 2
        self.battery += battery_increase
        #cuando la bateria este a 100%
        if self.battery > 100:
            self.battery = 100.0
        mode_info = " (modo alerta)" if self.alert_mode else ""
        print(f"{self.name} cargando{mode_info}... Batería: {self.battery}%")
#-------------------------------------------------------------------------------
#con modo alerta
    def toggle_alert(self):
        self.alert_mode = not self.alert_mode
        status = "on" if self.alert_mode else "off"
        print(f"modo alerta: {status}")

#muestra información incluyendo modo alerta
    def get_status(self):
        super().get_status()
        alert_status = "activado" if self.alert_mode else "desactivado"
        print(f"modo alerta: {alert_status}")

#---------------------Parte 3---------------------------------------------------
#clase robot team
class RobotTeam:
    def __init__(self, team_name):
        self.team_name = team_name
        self.robots = []  #lista vacía para almacenar robots
        self.missions_completed = 0  # Contador de misiones
        print(f"equipo {self.team_name} formado")

 #agregar robot al equipo
    def add_robot(self, robot):
        self.robots.append(robot)  #añade el robot a la lista
        print(f"{robot.name} se unió al equipo {self.team_name}")

    def __len__(self): #retorna el número de robots en el equipo
        return len(self.robots)

    def charge_all(self, time_minutes):#carga los robots en el equipo
        print(f"cargando equipo {self.team_name}")
        for robot in self.robots:
            robot.charge(time_minutes)

#estadisticas de los robots
    def team_status(self):
        print(f"estado del equipo {self.team_name.upper()}")
        print(f"total de robots: {len(self.robots)}")
        print(f"misiones completadas: {self.missions_completed}")
        print("--- Robots del equipo ---")

        #mostrar cada robot con número
        for i, robot in enumerate(self.robots):
            print(f"{i + 1}. {robot}")

    def complete_mission(self): #completa una misión
        self.missions_completed += 1
        print(f"misión completada total: {self.missions_completed}")

#-------------------------------------------------------------------------------
def main():
    robot1 = Robot("Alpha", "Megan-X1")
    robot2 = Robot("Beta", "Navigator-Z2")

    worker = WorkerRobot("Cargo", "Optimus-W3")

    guard = GuardRobot("Sentinel", "Guardian-G4")

    print("Moviendo robots")
    robot1.move(30, 45, 30)
    robot2.move(25, 180, 45)
    worker.move(20, 90, 60)
    guard.move(35, 270, 20)

    print("cargando robots")
    robot1.charge(20)
    robot2.charge(25)
    worker.charge(30)
    guard.charge(15)

    print("status individual")
    robot1.get_status()
    robot2.get_status()
    worker.get_status()
    guard.get_status()

    print("Funciones especiales")

    worker.load_cargo(50)
    worker.move(15, 135, 30)
    worker.unload_cargo()

    guard.toggle_alert()
    guard.move(40, 315, 15)
    guard.charge(10)
    guard.toggle_alert()

    print("CREAR EQUIPO:")

    team = RobotTeam("Alpha robots")

    print("agregando robots al equipo")
    team.add_robot(robot1)
    team.add_robot(robot2)
    team.add_robot(worker)
    team.add_robot(guard)

    print(f"el equipo tiene {len(team)} robots")

    print("Cargando todo el equipo")
    team.charge_all(10)

    team.team_status()

    print("Completando misiones")
    team.complete_mission()
    team.complete_mission()
    team.complete_mission()

    team.team_status()
    print("PRUEBA COMPLETADA EXITOSAMENTE")

    basic = Robot("Test1", "Basic")
    worker2 = WorkerRobot("Test2", "Worker")
    guard2 = GuardRobot("Test3", "Guard")

    basic.move(10, 0, 6)
    worker2.move(10, 0, 6)
    guard2.move(10, 0, 6)

    print("- Mismo método, diferentes comportamientos")
    robots = [basic, worker2, guard2]

    worker2.load_cargo(20)
    guard2.toggle_alert()

    for robot in robots:
        robot.move(15, 45, 12)

    print("composicion:")
    team2 = RobotTeam("test team")
    for robot in robots:
        team2.add_robot(robot)

    print(f"el equipo tiene {len(team2)} robots")
    team2.team_status()


if __name__ == "__main__":
    main()
```
